```python
midexam = pd.read_csv('data/midexam.csv')
finalexam = pd.read_csv('data/finalexam.csv')
midexam_r = pd.read_csv('data/midexam_r.csv')
finalexam_r = pd.read_csv('data/finalexam_r.csv')

# finalexam = finalexam.drop('Total', axis=1) # drop the column with label Total
# finalexam = finalexam.drop('№', axis=1) # drop the column with label №
# finalexam = finalexam.drop('Name', axis=1) # drop the column with label №


columns_to_remove = ['Total', '№', 'Name']
def drop_columns(df, columns_to_drop):
    return df.drop(columns_to_drop, axis=1)

midexam = drop_columns(midexam, columns_to_remove)
finalexam = drop_columns(finalexam, columns_to_remove)
midexam_r = drop_columns(midexam_r, columns_to_remove)
finalexam_r = drop_columns(finalexam_r, columns_to_remove)
```

```python
def encrypt_string(encr):
    encrypted_string = ""
    for char in encr:
        ascii_code = ord(char)
        encrypted_ascii_code = ascii_code + 1
        encrypted_char = chr(encrypted_ascii_code)
        encrypted_string += encrypted_char
    return encrypted_string

# Криптираме колоната "User" и съхраняваме резултата в нова колона "EncryptedUser"

midexam['User'] = midexam['User'].apply(encrypt_string)
finalexam['User'] = finalexam['User'].apply(encrypt_string)
midexam_r['User'] = midexam_r['User'].apply(encrypt_string)
finalexam_r['User'] = finalexam_r['User'].apply(encrypt_string)
```

```python
finalexam
```

```python
# #Unpivot Data
# finalexam = finalexam.melt(id_vars=['User'], value_vars=finalexam.columns[3:], var_name='Task', value_name='Points')
# finalexam['Task Number'] = finalexam['Task'].str.slice(0, 2) # extract the first 2 characters from Task
# finalexam = finalexam.replace('-', pd.NA) # replace '-' with NaN
# finalexam = finalexam.dropna() # drop rows with NaN
# finalexam = finalexam.reindex(columns=['User', 'Task Number', 'Task', 'Points']) # rearrange the columns


def unpivot_data(unp):
    unp = unp.melt(id_vars=['User'], value_vars=unp.columns[3:], var_name='Task', value_name='Points')
    unp['Task Number'] = unp['Task'].str.slice(0, 2) # extract the first 2 characters from Task
    unp = unp.replace('-', pd.NA) # replace '-' with NaN
    unp = unp.dropna() # drop rows with NaN
    unp = unp.reindex(columns=['User', 'Task Number', 'Task', 'Points']) # rearrange the columns
    return unp

# Извикване на функцията и промяна на DataFrame finalexam

midexam = unpivot_data(midexam)
finalexam = unpivot_data(finalexam)
midexam_r = unpivot_data(midexam_r)
finalexam_r = unpivot_data(finalexam_r)
```

```python
finalexam
```

```python
# Създаване на колоната Exam за всеки дейтафрейм
midexam['Exam'] = 'midexam'
finalexam['Exam'] = 'finalexam'
midexam_r['Exam'] = 'midexam_r'
finalexam_r['Exam'] = 'finalexam_r'

# Обединяване на дейтафреймовете в all_exams

all_exams = pd.concat([midexam, finalexam, midexam_r, finalexam_r], ignore_index=True)

all_exams['Points'] = all_exams['Points'].astype(int)
# Опционално: Реорганизация на колоните в желания ред
all_exams = all_exams.reindex(columns=['Exam', 'User', 'Task Number', 'Task', 'Points'])
```

```python
all_exams
```

```python
exam_avg = all_exams.pivot_table(index='Exam', columns='Task Number', values='Points', aggfunc='mean')
exam_avg = exam_avg.applymap(lambda x: '{:.2f}'.format(x))
```

```python
exam_avg
```

```python
# Преобразуване на exam_avg в числова матрица
exam_matrix = exam_avg.applymap(float).values

# Създаване на heatmap с imshow()
plt.imshow(exam_matrix, cmap='viridis', interpolation='nearest')

# Добавяне на цветна легенда
plt.colorbar()

# Добавяне на текст върху всеки квадрат
for i in range(len(exam_avg.index)):
    for j in range(len(exam_avg.columns)):
        plt.text(j, i, f'{exam_matrix[i, j]:.2f}', ha='center', va='center', color='black')

# Заглавия на осите
plt.xlabel('Task Number')
plt.ylabel('Exam')

# Заглавие на графиката
plt.title('Average Points per Task Number and Exam')

# Показване на графиката
plt.show()
```

```python
exam_matrix = exam_avg.applymap(float).values

plt.imshow(exam_matrix, cmap='autumn', interpolation='nearest')

plt.colorbar()

for i in range(len(exam_avg.index)):
    for j in range(len(exam_avg.columns)):
        plt.text(j, i, f'{exam_matrix[i, j]:.2f}', ha='center', va='center', color='black')

plt.xlabel('Task Number')
plt.ylabel('Exam')

plt.yticks(range(len(exam_avg.index)), exam_avg.index)
plt.xticks(range(len(exam_avg.columns)), ['01', '02', '03']) 
plt.title('Average Points per Task Number and Exam')

plt.show()
```

```python
exam_avg = exam_avg.applymap(lambda x: pd.to_numeric(x, errors='coerce'))

# Построяване на стълбовата диаграма
ax = exam_avg.plot(kind='bar', cmap='viridis')
# plt.legend(title='Task Number', loc='lower right')
plt.legend(title='Task Number', bbox_to_anchor=(1, 1))
plt.xlabel('Exam')
plt.ylabel('Average Points')
plt.title('Average Points per Task Number and Exam')
plt.show()
```

```python
# Групиране по "Task Number" и изчисляване на средната стойност за "Points"
finalexam['Points'] = finalexam['Points'].astype(int) # convert Points to integer
average_points_by_task = finalexam.groupby('Task Number')['Points'].mean()


# Плотиране на средните стойности
ax = average_points_by_task.plot(kind='bar')
plt.xlabel('Task Number')
plt.ylabel('Average Points')
plt.title('Average Points by Task Number')
ax.set_ylim(77, 84)

for i, v in enumerate(average_points_by_task):
    ax.text(i, v + 1, str(round(v, 2)), ha='center', va='bottom')

plt.show()
```

```python
counts = finalexam['Task'].value_counts().reset_index() # count the occurrences of each Task and reset the index
counts.columns = ['Task', 'Count'] # rename the columns
```

```python
counts
```

```python
pl = finalexam.groupby('User')['Points'].sum().reset_index() #Average Points by Task
pl = pl.sort_values(by='Points', ascending=False) # sort by Points in descending order
```

```python
pl
```

```python
# Списък с границите на групите
bins = [-1,0, 100, 200, 250, 299, 300]

# Групиране по стойностите на Points, използвайки границите на групите
grouped_data = pl.groupby(pd.cut(pl['Points'], bins=bins)).count()

# Създаване на нов дейтафрейм с резултатите
result_df = pd.DataFrame({
    'Interval': [f'{left+1}-{right}' for left, right in zip(bins[:-1], bins[1:])],
    'Count': grouped_data['User']
})
print(result_df)
```

```python
result_df
```

```python
plt.bar(result_df['Interval'], result_df['Count'])
plt.xlabel('Interval')
plt.ylabel('Count')
plt.title('Count of Users in Each Interval')
plt.xticks(rotation=45)  # Завъртане на надписите по осата X, за да са по-четливи

plt.show()
```
